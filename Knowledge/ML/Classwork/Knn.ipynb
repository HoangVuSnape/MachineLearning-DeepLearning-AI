```python
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

# Dữ liệu huấn luyện
X_train = np.array([
    [5, 45],  # Chiều cao, tuổi
    [5.11, 26],
    [5.6, 30],
    [5.9, 34],
    [4.8, 40],
    [5.8, 36],
    [5.3, 19],
    [5.8, 28],
    [5.5, 23],
    [5.6, 32]
])

# Nhãn của dữ liệu huấn luyện
y_train = np.array(['H', 'L', 'M', 'M', 'H', 'M', 'L', 'M', 'L', 'M'])

# Điểm mới cần dự đoán
new_data_point = np.array([[5.5, 38]])  # Chiều cao, tuổi của điểm mới

# Tính khoảng cách từ điểm mới đến các điểm huấn luyện
distances = np.sqrt(np.sum((X_train - new_data_point) ** 2, axis=1))

# Khởi tạo mô hình KNN với k=3 và dự đoán
knn_model = KNeighborsClassifier(n_neighbors=3)
knn_model.fit(X_train, y_train)
nearest_neighbors = knn_model.kneighbors(new_data_point, return_distance=True)

# Visualize distances
plt.figure(figsize=(10, 6))
plt.plot(range(len(X_train)), sorted(distances), marker='o')
plt.xlabel('Index of Training Instances')
plt.ylabel('Distance to New Point')
plt.title('Distances to Nearest Neighbors')
plt.xticks(range(len(X_train)))
plt.grid(True)
plt.show()

# Print distances and corresponding classes of nearest neighbors
print("Distances to nearest neighbors:", nearest_neighbors[0])
print("Indices of nearest neighbors:", nearest_neighbors[1])
print("Classes of nearest neighbors:", y_train[nearest_neighbors[1]])
```
