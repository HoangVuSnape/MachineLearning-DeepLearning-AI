### Pyplot API and object-oriented OOP API
* pyplot api - quickly
* oo api - advanced


```python
x = np.linspace(0, 10, 1000)
plt.plot(x, np.sin(x), color = "blue", linestyle = "dashed", label = "sin(x)")
plt.plot(x, np.cos(x), color = "red", label = "cos(x)")

plt.title("Sin(x) and cos(X)")
plt.xlabel("bien x") 
plt.ylabel("sin(x)") 
# plt.xlim(0, 10) # limitation of axis x
# plt.ylim(-.7, .6) # limitation of axis x
# plt.axis([.5, 10, -0.7, 0.6])  #plt.axis(xmin, xmax, ymin, ymax)
plt.axis("tight")
plt.axis("equal")


plt.legend(); # it help to display label of 2 plot
```

Object-Oriented API 

```python
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]
fig, ax = plt.subplots(figsize = (5,5)) # figure size = wid and height of plt

#plot of data
ax.plot(x, y);

ax.set(title= "A simple plot", xlabel="x-axis", ylabel= "y-axis");
```

```python
names = ['vu', 'mai', 'hang']
marks = [22, 28, 26]

plt.figure(figsize=(9,3))

plt.subplot(131)
plt.bar(names, marks)

plt.subplot(132)
plt.scatter(names, marks)

plt.subplot(133)
plt.plot(names, marks, marker ='o')

plt.suptitle('dates of birth')
plt.show()
```

```python
#00 api - distribution
rng = np.random.RandomState(0)

x = rng.randn(100)
y = rng.randn(100)
sizes = 1000 * rng.randn(100)
colors = rng.randn(100)


fig, ax = plt.subplots()
img = ax.scatter(x, y, s = sizes, c=colors, cmap= 'viridis', alpha = 0.3); # cmap - colors --- alpha - do  trong suot.
fig.colorbar(img)
```

```python
# prepare data
soft_drink_prices = {
    "Coke": 10,
    "Pessi": 12,
    "Sprite": 8
}

fig, ax = plt.subplots()
ax.bar(soft_drink_prices.keys(), soft_drink_prices.values())
ax.set(title= "Back hoa xanh soft drink prices", ylabel= "prices($)")
```

```python
fig, ax = plt.subplots()
ax.barh(list(soft_drink_prices.keys()), list(soft_drink_prices.values()))
```

```python
#Option 2
fig, ax = plt.subplots(nrows= 2,
                       ncols= 2,
                       figsize = (10,5)  
)

ax[0,0].plot(x, x/2);
ax[0,0].plot(x, x*2);

ax[0, 1].scatter(np.random.random(10), np.random.random(10));

ax[1, 0].bar(soft_drink_prices.keys(), soft_drink_prices.values());

ax[1, 1].hist(student_height, bins= 30);
```
